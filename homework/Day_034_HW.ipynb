假設我們資料中類別的數量並不均衡，在評估準確率時可能會有所偏頗，試著切分出 y_test 中，0 類別與 1 類別的數量是一樣的 (亦即 y_test 的類別是均衡的)

```python
import numpy as np
X = np.arange(1000).reshape(200, 5)
y = np.zeros(200)
y[:40] = 1
```

```python
from sklearn.model_selection import train_test_split, KFold
import numpy as np
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

可以看見 y 類別中，有 160 個 類別 0，40 個 類別 1 ，請試著使用 train_test_split 函數，切分出 y_test 中能各有 10 筆類別 0 與 10 筆類別 1 。(HINT: 參考函數中的 test_size，可針對不同類別各自作切分後再合併)

```python
kf = KFold(n_splits=10)
i = 0
#print(X[0:8])
#print(y[0:40])
for train_index, test_index in kf.split(X):
    i +=1 
    #print(train_index)
    #%print(test_index)
    x_train_1 ,x_test_1 , y_train_1   , y_test_1 = train_test_split(X[0:40] , y[0:40] )
    print("FOLD {}: ".format(i))
    print("X_test_1: ", x_test_1)
    print("X_train_1: ", x_train_1)
    print("Y_test_1: ", y_test_1)
    print("Y_train_1: ", y_train_1)
    print("-"*30)
    x_train_2 ,  x_test_2,y_train_2  , y_test_2 = train_test_split(X[40:] , y[40:] ,test_size=10)

    print("FOLD {}: ".format(i))
    print("X_test_2: ", x_test_2)
    print("X_train_2: ", x_train_2)
    print("Y_test_2: ", y_test_2)
    print("Y_train_2: ", y_train_2)
    print("-"*30)
```
